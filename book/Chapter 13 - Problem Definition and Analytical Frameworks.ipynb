## 13.1 Business Problems vs Analytical Problems

A **business problem** is usually written in business language and is often broad or vague.

Examples:
- "Sales are down"
- "Customers are leaving"
- "Marketing feels less effective"

An **analytical problem** is a question that data can help answer. It is specific enough that you can:
- define what to measure
- identify the needed data
- choose an approach (descriptive analysis, segmentation, forecasting, etc.)

### Example Conversions

| Business Problem | Analytical Problem |
|------------------|-------------------|
| Sales are down | Which products and channels saw the largest revenue decline in the last 8 weeks? |
| Customers are leaving | Did retention drop, and which customer segments have the highest churn? |
| Marketing feels less effective | How has cost-per-acquisition changed by channel over the last quarter? |

### A Beginner Template

When you get a vague request, try rewriting it as:

> **How does [Outcome] change across [Groups] over [Time], and what factors might explain it?**

Where:
- **Outcome** = the metric (revenue, conversion rate, churn, etc.)
- **Groups** = channel, product, segment, region, device
- **Time** = a clear comparison window

---

> ðŸ’¡ **Tip:** If you can't name the outcome and how you'll measure it, the problem is not analytics-ready yet. Go back to stakeholders for clarification.

> âš ï¸ **Common Mistake:** Jumping into analysis without defining the problem clearly. This often leads to "interesting findings" that don't answer anyone's question.

````python
# A small helper to practice translating a business request into an analytics question

def build_analytics_question(outcome: str, groups: list[str], time_window: str, context: str | None = None) -> str:
    groups_text = ', '.join(groups)
    question = f'How does {outcome} change across {groups_text} over {time_window}?'
    if context:
        question += f' ({context})'
    return question

example = build_analytics_question(
    outcome='conversion rate',
    groups=['marketing channel', 'device type'],
    time_window='the last 12 weeks',
    context='focus on new customers'
)
print(example)
````

## 13.2 Defining Objectives and Success Criteria

A good analytics objective answers: **what decision will someone make using this analysis?**

### Objective vs Success Criteria

| Term | Definition | Example |
|------|------------|---------|
| **Objective** | The direction (what you want to achieve) | Increase online revenue |
| **Success Criteria** | The measurable target (how you know you succeeded) | Increase weekly revenue by 10% within 8 weeks without decreasing profit margin |

### SMART Goals Framework

SMART is a widely-used framework for writing good objectives:

| Letter | Meaning | Question to Ask |
|--------|---------|-----------------|
| **S** | Specific | What exactly do we want to achieve? |
| **M** | Measurable | How will we measure progress? |
| **A** | Achievable | Is this realistic given our resources? |
| **R** | Relevant | Does this align with business priorities? |
| **T** | Time-bound | When should this be achieved? |

---

> âš ï¸ **Common Mistake:** Writing success criteria like "improve satisfaction" without defining the metric (survey score, NPS, complaint rate, etc.). Always specify *how* you will measure success.

> ðŸ’¡ **Tip:** Ask "How would I know if we failed?" If you can't answer that, your success criteria isn't specific enough.

## 13.3 Identifying Stakeholders

A **stakeholder** is anyone who:
- Requests the analysis
- Uses the results to make decisions
- Owns the data you need
- Is impacted by the decisions made

### Stakeholder Analysis Framework

For each stakeholder, document:

| Attribute | Description |
|-----------|-------------|
| **Role/Title** | Who they are in the organization |
| **Decision** | What decision they will make based on your analysis |
| **Key Questions** | What they want to know |
| **Constraints/Concerns** | Budget, privacy, deadlines, risk tolerance |
| **Data They Own** | What data sources they control |

---

> ðŸ’¡ **Tip:** Align on definitions early! Create a small glossary of terms (e.g., "What counts as a 'customer'? Someone who registered, or only those who purchased?"). Misaligned definitions cause many analytics problems.

> âš ï¸ **Warning:** Different stakeholders may have conflicting goals. Marketing may want more spend, while Finance wants cost reduction. Your job is to surface these trade-offs, not hide them.

## 13.4 KPI and Metric Definition

### Metrics vs KPIs

| Term | Definition | Example |
|------|------------|---------|
| **Metric** | Any measurable quantity | Page views, session duration, clicks |
| **KPI** (Key Performance Indicator) | A metric that directly reflects progress toward the objective | Conversion rate, revenue, customer lifetime value |

> Not all metrics are KPIs, but all KPIs are metrics.

### KPI Specification Template

Define each KPI like a contract to avoid confusion:

| Attribute | Description | Example |
|-----------|-------------|---------|
| **Name** | Clear, unambiguous name | Weekly Conversion Rate |
| **Formula** | Exact calculation | (Orders / Visits) Ã— 100 |
| **Unit** | Percentage, currency, count | Percentage (%) |
| **Grain** | Level of detail | Weekly, by channel |
| **Filters** | Exclusions/inclusions | Exclude internal traffic |
| **Data Source** | Where the data comes from | Google Analytics, Sales DB |

---

> âš ï¸ **Warning:** A KPI can improve while the business gets worse! For example, conversion rate might increase because you're attracting fewer but more qualified visitors, while total revenue drops. Always track a *balanced set* of KPIs.

> ðŸ’¡ **Tip:** Include at least one "guardrail" metric that ensures you're not optimizing one thing at the expense of something important (e.g., track profit margin alongside revenue).

````python
# Synthetic dataset for KPI practice
rng = np.random.default_rng(42)
weeks = pd.date_range('2025-09-01', periods=12, freq='W-MON')
channels = ['Search', 'Social', 'Email']

rows = []
for week in weeks:
    for channel in channels:
        visits = int(max(rng.normal(12000, 1500), 1000))
        base_cr = {'Search': 0.030, 'Social': 0.018, 'Email': 0.045}[channel]
        conv_rate = float(np.clip(rng.normal(base_cr, 0.004), 0.001, 0.20))
        orders = int(round(visits * conv_rate))
        base_aov = {'Search': 68, 'Social': 60, 'Email': 75}[channel]
        aov = float(max(rng.normal(base_aov, 7), 5))
        revenue = orders * aov
        base_spend = {'Search': 14000, 'Social': 9000, 'Email': 2000}[channel]
        ad_spend = float(max(rng.normal(base_spend, 1800), 0))
        rows.append({'week': week, 'channel': channel, 'visits': visits, 'orders': orders, 'revenue': revenue, 'ad_spend': ad_spend})

df = pd.DataFrame(rows)
df.head()
````

````python
df_kpis = df.copy()
df_kpis['conversion_rate'] = df_kpis['orders'] / df_kpis['visits']
df_kpis['roas'] = df_kpis['revenue'] / df_kpis['ad_spend'].replace(0, np.nan)
df_kpis.head()
````

### Visual Example
Plot conversion rate by channel over time.

````python
pivot = df_kpis.pivot_table(index='week', columns='channel', values='conversion_rate')
ax = pivot.plot(marker='o', figsize=(10, 4))
ax.set_title('Weekly Conversion Rate by Channel')
ax.set_xlabel('Week')
ax.set_ylabel('Conversion Rate')
ax.grid(True, alpha=0.3)
plt.show()
````

## 13.5 Hypothesis-Driven Analysis

A **hypothesis** is a clear, testable statement about what you expect to find in the data.

### Why Use Hypotheses?

1. **Focus:** Prevents aimless data exploration
2. **Efficiency:** Guides what data to collect and analyze
3. **Objectivity:** Reduces confirmation bias when stated upfront
4. **Communication:** Makes findings easier to explain

### Hypothesis Structure

A good hypothesis has this structure:

> **[Group A] has [higher/lower] [metric] than [Group B] because [reason].**

Examples:
- "Email channel has a higher conversion rate than Social because email subscribers are more engaged."
- "Weekend orders have higher average order value than weekday orders because customers have more time to browse."
- "Mobile users have a lower checkout completion rate than desktop users because of UX friction."

### Null vs Alternative Hypothesis

| Type | Definition | Example |
|------|------------|---------|
| **Null Hypothesis (Hâ‚€)** | No difference exists | Email conversion = Social conversion |
| **Alternative Hypothesis (Hâ‚)** | A difference exists | Email conversion â‰  Social conversion |

---

> ðŸ’¡ **Tip:** Write your hypothesis *before* looking at the data. This prevents "HARKing" (Hypothesizing After Results are Known), which leads to false discoveries.

> âš ï¸ **Warning:** If you change your hypothesis after seeing results, be transparent about it. Label it as "exploratory finding" rather than "confirmed hypothesis."

````python
email_mean = df_kpis.loc[df_kpis['channel'] == 'Email', 'conversion_rate'].mean()
social_mean = df_kpis.loc[df_kpis['channel'] == 'Social', 'conversion_rate'].mean()
print(f'Email mean conversion rate:  {email_mean:.3%}')
print(f'Social mean conversion rate: {social_mean:.3%}')
print('Supports hypothesis (descriptively)?', email_mean > social_mean)
````

## 13.6 Analytics Project Planning

A well-structured project plan keeps everyone aligned and helps you avoid common pitfalls.

### Key Components of an Analytics Project Plan

| Component | Description | Questions to Answer |
|-----------|-------------|---------------------|
| **Problem Statement** | Clear description of the issue | What is happening? Why does it matter? |
| **Objective** | What you want to achieve | What decision will this analysis enable? |
| **Success Criteria** | How you'll measure success | What number/outcome means we succeeded? |
| **Stakeholders** | Who is involved | Who needs this? Who has the data? |
| **KPIs** | Metrics to track | What will we measure? How? |
| **Data Requirements** | What data you need | What sources? What time range? What granularity? |
| **Deliverables** | What you'll produce | Report? Dashboard? Presentation? |
| **Timeline** | When things happen | Milestones and deadlines |
| **Risks/Assumptions** | What could go wrong | Data quality? Scope creep? |

### Project Timeline Template

A typical analytics project follows these phases:

```
Week 1: Define â†’ Confirm metrics, stakeholders, data sources
Week 2: Collect â†’ Gather and validate data
Week 3: Analyze â†’ Perform analysis, test hypotheses  
Week 4: Report â†’ Present findings, get feedback
Week 5: Refine â†’ Iterate based on feedback
```

---

> ðŸ’¡ **Tip:** Schedule early checkpoints with stakeholders (e.g., at 30% completion). This catches misalignments before you've invested too much work.

> âš ï¸ **Common Mistake:** Underestimating data collection and cleaning time. A good rule of thumb: data preparation often takes 60-80% of project time.

````python
project_plan = {
    'Problem statement': 'Social channel performance appears weaker. We need to decide whether to shift budget or improve landing pages.',
    'Objective': 'Improve marketing efficiency',
    'Success criteria': 'Increase ROAS by 15% within 8 weeks without decreasing total revenue',
    'Stakeholders': ['Marketing Manager', 'Finance', 'Web/Product'],
    'KPIs': {'ROAS': 'revenue / ad_spend', 'Conversion rate': 'orders / visits', 'Revenue': 'sum(order_value)'},
    'Data requirements': ['Visits by channel', 'Orders and revenue', 'Ad spend by channel'],
    'Deliverables': ['Notebook with findings', '1-page summary'],
    'Timeline': ['Week 1: confirm metrics + pull data', 'Week 2: analysis + checkpoint', 'Week 3: recommendations'],
    'Risks/assumptions': ['Attribution may be imperfect', 'Tracking is correct'],
}
project_plan
````

````python
# Visual: Simple Gantt Chart for Analytics Project Timeline

fig, ax = plt.subplots(figsize=(12, 5))

# Define project phases
phases = [
    {'name': 'Problem Definition', 'start': 0, 'duration': 1, 'color': '#3498db'},
    {'name': 'Data Collection', 'start': 1, 'duration': 1.5, 'color': '#2ecc71'},
    {'name': 'Data Cleaning', 'start': 2, 'duration': 2, 'color': '#f39c12'},
    {'name': 'Analysis', 'start': 3.5, 'duration': 2, 'color': '#9b59b6'},
    {'name': 'Stakeholder Review', 'start': 4.5, 'duration': 0.5, 'color': '#e74c3c'},
    {'name': 'Reporting', 'start': 5, 'duration': 1, 'color': '#1abc9c'},
]

# Plot Gantt bars
for i, phase in enumerate(phases):
    ax.barh(i, phase['duration'], left=phase['start'], height=0.6, 
            color=phase['color'], alpha=0.8, edgecolor='black')
    ax.text(phase['start'] + phase['duration']/2, i, phase['name'], 
            ha='center', va='center', fontsize=10, fontweight='bold', color='white')

ax.set_yticks(range(len(phases)))
ax.set_yticklabels([f"Phase {i+1}" for i in range(len(phases))])
ax.set_xlabel('Weeks', fontsize=12)
ax.set_title('Typical Analytics Project Timeline', fontsize=14)
ax.set_xlim(0, 7)
ax.grid(True, axis='x', alpha=0.3)
ax.invert_yaxis()

plt.tight_layout()
plt.show()

print("\nðŸ“‹ Key Milestones:")
print("â€¢ Week 1: Problem definition complete, stakeholders aligned")
print("â€¢ Week 2.5: Data collected and validated")
print("â€¢ Week 4: Initial findings ready for review")
print("â€¢ Week 5: Stakeholder checkpoint")
print("â€¢ Week 6: Final report delivered")
````

## Summary / Key Takeaways

### What You Learned

1. **Business vs Analytical Problems**
   - Business problems are broad and vague; analytical problems are specific and measurable
   - Use the template: "How does [Outcome] change across [Groups] over [Time]?"

2. **Objectives and Success Criteria**
   - Objectives define direction; success criteria define measurable targets
   - Use the SMART framework: Specific, Measurable, Achievable, Relevant, Time-bound

3. **Stakeholder Identification**
   - Map stakeholders by their decisions, questions, and constraints
   - Align on definitions early to avoid miscommunication

4. **KPI Definition**
   - KPIs are metrics directly tied to objectives
   - Define KPIs precisely: name, formula, unit, grain, filters, data source
   - Track balanced sets of KPIs to avoid unintended consequences

5. **Hypothesis-Driven Analysis**
   - Formulate testable hypotheses before analyzing data
   - Structure: "[Group A] has [higher/lower] [metric] than [Group B]"

6. **Project Planning**
   - Document problem, objective, stakeholders, KPIs, data needs, timeline, and risks
   - Schedule early checkpoints with stakeholders
